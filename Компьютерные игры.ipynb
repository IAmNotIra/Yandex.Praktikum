## Подготовка данных

Для лаконичности приведём названия колонок к нижнему регистру.

```python
df.columns = df.columns.str.lower()
```

### Обработка пропущенных значений и дубликатов

* Удалим строки с пропусками в названии игр, годах релиза и жанре, т.к. эти данные необходимо запрашивать у разработчиков, а не заменять, например, средним или самым часто встречающимся значением. 
* Слишком большое количество пропущенных данных в столбцах оценок критиков, игроков и рейтинге. Если заполнить их, то они сильно повлияют на распределения в дальнейшем и исказят статистику. Пока оставим.
* В столбце user_score есть значения tbd (to be determined - отсутсвие оценки на момент выгрузки фрейма). Поменяем его на NaN.

```python
df.dropna(subset=['name', 'year_of_release', 'genre'], axis=0, inplace=True)
df.head()
```

```python
# КОД РЕВЬЮЕРА 1

df.name.str.extract('(\d+)')
```

```python
# КОД РЕВЬЮЕРА 2

df['name'].str.extract(r'(\d\d\d\d).*')
```

Так как в оценках и рейтинге много пропусков, посмотрим, есть ли дубликаты по названию, платформе, году релиза, и жанру.

```python
df.loc[df.duplicated(subset = ['name', 'platform', 'year_of_release', 'genre'], keep=False)]
```

Всего один и отличается только продажами. Удаляем.

```python
df.drop_duplicates(subset = ['name', 'platform', 'year_of_release', 'genre']).reset_index(drop=True)
df.head()
```

Заменим tbd на NaN, а пропущенные значения в рейтинге на "NR" (no rating).

```python
df['user_score'].replace('tbd', np.nan, inplace=True)
df['rating'] = df['rating'].fillna('NR')
```

### Изменение типа данных

```python
df['user_score'] = pd.to_numeric(df['user_score'], downcast='integer')
df['critic_score'] = pd.to_numeric(df['critic_score'], downcast='integer') # из-за пропусков целые числа не получить
df['year_of_release'] = pd.to_numeric(df['year_of_release'], downcast='integer')
```

```python
df.info()
```

### Добавление результатов
Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец.

```python
df['world_sales'] = (df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales'])
df.insert(df.columns.get_loc('other_sales')+1, 'world_sales',df.pop('world_sales'))
df.head()
```

### Вывод
На этапе подготовки мы выполнили:
* Название столбцов приведено к нижнему регистру.
* Удалены строки с пропусками в нужных данных (название, год релиза, жанр).
* Удалены дубликаты.
* Изменён тип данных.
* Добавлен столбец суммарных продаж.  

Данные готовы к анализу.

Тперь посмотрим, как менялись продажи по платформам.

```python
platform_table = pd.pivot_table(df, 
                                index='platform', 
                                values='world_sales', 
                                aggfunc='sum').sort_values('world_sales', 
                                                           ascending=False)
platform_table.plot(kind='bar', 
                    y='world_sales', 
                    grid=True ,
                    legend=False, 
                    title='Количество продаж по платформам', 
                    figsize=(12, 6));
```

Сложно будет выделить топ-5 платформ, видим разницу в 2 раза между PS и PS4. Но мы знаем, что последней крупной игрой, выпущенной для PS была FIFA Soccer 2004. Поэтому возьмём пятёрку лидеров.

```python
top = platform_table.head(5)
top
```

И сравним на линейном графике.

```python
def z(name, data):
    return data.query(
        'platform == @name').pivot_table(
        index='year_of_release', 
        values='world_sales', 
        aggfunc='sum').sort_values(
        'year_of_release', 
        ascending=False)   
```

```python
for i in list(top.index):
    z(i, df)['world_sales'].plot(
        figsize =(12, 6), 
        grid=True, 
        label=i, 
        title='Распределение продаж для популярных платформ по годам').set(
        xlabel='Годы игровых релизов', 
        ylabel='Продажи')
    plt.legend()
```

Видим, что в среднем жизненный цикл (от внедрения на рынок до полного спада популярности) платформ составляет 10 лет. Самое большое количество игр было выпущено для PS2 и Wii.

```python
def regional_top(column, region):
    return data.pivot_table(index=column, 
                            values=region, 
                            aggfunc='sum').sort_values(by=region, 
                                                       ascending=False).head()
```

```python
def pie(data, name, region):
    (data.plot.pie(
        y=region, 
        legend=False,
        autopct='%1.0f%%', 
        cmap='Blues',
        figsize=(12,6), title=name))
    plt.show()
```

## Проверка гипотез

###  Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Так как нулевая гипотеза всегда формулируется так, чтобы использовать знак равенства, а альтернативная гипотеза  является отрицанием нулевой, мы получим:  


***H$_0$***  — средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу платформы PC.  
***H$_1$*** — средние пользовательские рейтинги платформ Xbox One и PC различаются.  

Считать равными дисперсии не будем, укажем это в параметре equal_var.  

Укажем критический уровень статистической значимости 1%.  
$\alpha$ = 0.01


```python
pc = data[data['platform'] == "PC"] 
xone = data[data['platform'] == "XOne"]

alpha = .01
results = st.ttest_ind(pc['user_score'], xone['user_score'], nan_policy='omit', equal_var=False)
print('значение p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем Нулевую гипотезу')
else:
    print('Не получилось отвергнуть Нулевую гипотезу')
```

```python
display(pc['user_score'].describe(), xone['user_score'].describe())
```

Cреднее и медиана рейтингов похожи друг на друга, что и подтверждает нулевая гипотеза.

### Средние пользовательские рейтинги жанров Action и Sports разные

***H$_0$***  — средний пользовательский рейтинг жанра Action равен среднему пользовательскому рейтингу жанра Sport.  
***H$_1$*** — Средние пользовательские рейтинги жанров Action и Sports разные. 

Считать равными дисперсии не будем, укажем это в параметре equal_var.  

Укажем критический уровень статистической значимости 1%.  
$\alpha$ = 0.01

```python
action = data[data['genre'] == "Action"]
sport = data[data['genre'] == "Sports"]

alpha = .01
results = st.ttest_ind(action['user_score'], sport['user_score'], nan_policy='omit', equal_var=False)
print('значение p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем Нулевую гипотезу')
else:
    print('Не получилось отвергнуть Нулевую гипотезу')
```

```python
display(action['user_score'].describe(), sport['user_score'].describe())
```

Видим существенное различие среднего и медианного значений. Тест Стюдента подтвердил, что данные различаются.
